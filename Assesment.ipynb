```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("DataAnalyst_Assesment_Dataset.xlsx")
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
# Fill missing numerical values with median
data.fillna(data.median(numeric_only=True), inplace=True)
```

```python
data.isnull().sum()
```

```python
# Check for duplicate records
duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Drop duplicates
data.drop_duplicates(inplace=True)
```

```python
# Convert booking dates to datetime format (modify column name accordingly)
data['Booking Date'] = pd.to_datetime(data['Booking Date'], errors='coerce')

# Extract year, month, and weekday for analysis
data['Year'] = data['Booking Date'].dt.year
data['Month'] = data['Booking Date'].dt.month
data['Weekday'] = data['Booking Date'].dt.day_name()
```

```python
data
```

```python
#EDA & Visualization
#Booking Trends Over Time
import warnings
warnings.filterwarnings("ignore")
plt.figure(figsize=(10,5))
sns.countplot(data=data, x='Month', palette='coolwarm')
plt.title('Monthly Booking Trends')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
#Service Type Distribution
plt.figure(figsize=(8,5))
sns.countplot(data=data, x='Service Type', palette='viridis')
plt.title('Service Type Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by Service Name to find top revenue-generating services
top_services = data.groupby("Service Name")["Price"].sum().sort_values(ascending=False).head(10)

# Plot Top 10 Revenue-Generating Services
plt.figure(figsize=(12,5))
sns.barplot(x=top_services.index, y=top_services.values, palette="coolwarm")
plt.title("Top 10 Revenue-Generating Services")
plt.xlabel("Service Name")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
#Monthly Revenue Trends
# Group by month to calculate total revenue per month

monthly_revenue = data.groupby("Month")["Price"].sum()

# Plot Monthly Revenue Trends
plt.figure(figsize=(10,5))
sns.lineplot(x=monthly_revenue.index, y=monthly_revenue.values, marker="o", color="b")
plt.title("Monthly Revenue Trends")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(range(1,13))
plt.grid()
plt.show()
```

```python
#Booking Type vs. Revenue

# Total revenue by booking type
booking_revenue = data.groupby("Booking Type")["Price"].sum()

# Plot Revenue by Booking Type
plt.figure(figsize=(8,5))
sns.barplot(x=booking_revenue.index, y=booking_revenue.values, palette="viridis")
plt.title("Revenue by Booking Type")
plt.xlabel("Booking Type")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
cleaned_file = "DataAnalyst_Assesment_Dataset.csv"
data.to_csv(cleaned_file, index=False)
print(f"Cleaned dataset saved as {cleaned_file}")
```
